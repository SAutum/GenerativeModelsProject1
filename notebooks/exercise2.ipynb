```python
import torch
from torch.utils.data.dataloader import DataLoader

from data_utils.BrainDataset import BrainDataset
from data_utils.BrainSampler import BrainSampler
from data_utils import visualize_batch
```

```python
torch.random.manual_seed(0)
training_brains ={"B01", "B02", "B05", "B07"}
data = BrainDataset(training_brains)
sampler = BrainSampler(data, tile_size=64, weighting="by_pixel")
loader = DataLoader(data, sampler=sampler, batch_size=8)
batch = next(iter(loader))
visualize_batch(batch)
```
